```python
with open('../data/legislators.json') as f:
    data = json.loads(f.read())
    
df = pandas.json_normalize(data, "terms", [['name', 'official_full'], 'bio', 'id'])
df.head()
```

```python
df = df.set_index(['state', 'district'])
df.head()
```

```python
# Convert start and end columns to dates
df[['start', 'end']] = df[['start', 'end']].apply(lambda x: pandas.to_datetime(x).dt.date)
```

```python
# Filter out old terms, just get the current ones
df = df[(df['start'] <= active_date) & (df['end'] >= active_date)]
df
```

```python
# Filter out senators and just get representatives
df = df[(df['type'] == 'rep')]
df
```
